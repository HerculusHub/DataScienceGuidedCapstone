```python
# Imports
#Import pandas and seaborn 

import pandas as pd
import seaborn as sns
import os
```

```python
#Load The realtor-data
realtor_data = pd.read_csv('realtor-data.zip.csv')

#Call the info method on realtor-data to see a summary of the data
realtor_data.info()
```

```python
realtor_data.shape
```

```python
realtor_data2 = realtor_data.rename(index=str, columns={"acre_lot":"land_size"})
realtor_data2.head()
```

```python
realtor_data2.tail()
```

Description of the columns：
o	status (Housing status - a. ready for sale or b. ready to build)
o	bed (# of beds)
o	bath (# of bathrooms)
o	land_size (Property / Land size in acres，means the total land area)
o	city (city name)
o	state (state name)
o	zip_code (postal code of the area)
o	house_size (house area/size/living space in square feet，denotes the living space/building area)
o	prev_sold_date (Previously sold date)
o	price (Housing price, it is either the current listing price or recently sold price if the house is sold recently)

```python
realtor_data2_index_state= realtor_data2.set_index(['state']).groupby(['state']).mean()
print(realtor_data2_index_state)
```

```python
realtor_data2_index_city= realtor_data2.set_index(['state','city']).groupby(['state','city']).mean()
print(realtor_data2_index_city)
```

```python
realtor_data2_index_zipcode= realtor_data2.set_index(['state','city','zip_code']).groupby(['state','city','zip_code']).mean()
print(realtor_data2_index_zipcode)
```

```python
realtor_data2_drop=realtor_data2
realtor_data2_drop.dropna(subset =['zip_code', 'city'],inplace=True)
missing = pd.concat([realtor_data2_drop.isnull().sum(), 100 * realtor_data2_drop.isnull().mean()], axis=1)
missing.columns=['count','%']
missing.sort_values(by='%')
```
